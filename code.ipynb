```python
pgn = open('/Users/charleslefort/Etudes/Malte/201920_CPS3235_LEFORT_CHARLES_160475Z01117_assignment_code.zip/task2/KingBase2019-pgn/KingBase2019-A00-A39.pgn')


#variable for the lopp

i = 0
EloW = 0
EloL = 0
EloD = 0

#how many games we want to annalize
NbGames = 1000

while i < NbGames:
    #using the fonctionnality of pgn
    game = chess.pgn.read_game(pgn)
    
    if (game.headers["WhiteElo"] > game.headers["BlackElo"])&(game.headers["Result"] == "1-0") :
        EloW = EloW + 1    
        
    if (game.headers["WhiteElo"] < game.headers["BlackElo"])&(game.headers["Result"] == "0-1") :
        EloW = EloW + 1
        
    if (game.headers["WhiteElo"] < game.headers["BlackElo"])&(game.headers["Result"] == "1-0") :
        EloL = EloL + 1
        
    if (game.headers["WhiteElo"] > game.headers["BlackElo"])&(game.headers["Result"] == "0-1") :
        EloL = EloL + 1
     
    i = i + 1
    
    
EloD = NbGames - EloW - EloL

print(EloW)
print(EloL)
print(EloD)

    
```

```python
# visualisation with plt.bar

x = np.arange(3)
plt.bar(x, height= [EloW,EloL,EloD])
plt.xticks(x, ['EloW','EloL','other'])
plt.xlabel('Result per ELO')
plt.ylabel('number of games')
plt.title('statistic of winning by analyzing the ELO')
```
